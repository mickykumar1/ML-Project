```python
# Separating the columns as categorical and numerical
ordcols=['OverallQual','OverallCond','Alley','LotShape','LandContour','Utilities','LandSlope',
         'ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
         'HeatingQC','CentralAir','Electrical','KitchenQual','Functional','FireplaceQu','GarageQual',
         'GarageCond','PoolQC','PavedDrive']
catcols=traindf.select_dtypes('object').columns.to_list()
catcols=[i for i in catcols if i not in ordcols]
removelist=['Condition1','Condition2','RoofMatl','MiscFeature']
catcols=[i for i in catcols if i not in removelist]
numcols=traindf.select_dtypes('float64').columns.to_list()
catcols=catcols+['MSSubClass']
numcols=numcols+['LotArea','YearBuilt','YearRemodAdd','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF',
'TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath',
'BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd',
'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF','EnclosedPorch',
'3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']
target='SalePrice'
analysiscols=ordcols+catcols+numcols
len(analysiscols)
```

```python
def assignOrdvalues(df):
    Alley_St_dict={'nan':0 ,'Grvl':1,'Pave':2 }
    LotShape_dict={'Reg':3,'IR1':2,'IR2':1,'IR3':0}
    LandContour_dict={'Lvl':3,'Bnk':2,'HLS':1,'Low':0}
    Utilities_dict={'AllPub':3,'NoSewr':2,'NoSeWa':1,'ELO':0}
    LandSlope_dict={'Gtl':2,'Mod':1,'Sev':0}
    Exter_dict={'Ex':5,'Gd':4,'TA':3,'Fa':2,'Po':1,'nan':0} 
    BsmtExp_dict={'Gd':4,'Av':3,'Mn':2,'No':1,'nan':0}
    BsmtFinType1={'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,'nan':0}
    CentralAir_dict={'N':0,'P':1,'Y':2}
    Electrical_dict={'SBrkr':4,'FuseA':3,'FuseF':4,'FuseP':5,'Mixed':6}
    Functional_dict={'Typ':8,'Min1':7,'Min2':6,'Mod':5,'Maj1':4,'Maj2':3,'Sev':1,'Sal':0}
    df['Alley']=df.Alley.map(Alley_St_dict)
    df['LotShape']=df.LotShape.map(LotShape_dict)
    df['LandContour']=df.LandContour.map(LandContour_dict)
    df['Utilities']=df.Utilities.map(Utilities_dict)
    df['LandSlope']=df.LandSlope.map(LandSlope_dict)
    df['ExterQual']=df.ExterQual.map(Exter_dict)
    df['ExterCond']=df.ExterCond.map(Exter_dict)
    df['BsmtQual']=df.BsmtQual.map(Exter_dict)
    df['BsmtCond']=df.BsmtCond.map(Exter_dict)
    df['BsmtExposure']=df.BsmtExposure.map(BsmtExp_dict)
    df['BsmtFinType1']=df.BsmtFinType1.map(BsmtFinType1)
    df['BsmtFinType2']=df.BsmtFinType2.map(BsmtFinType1)
    df['HeatingQC']=df.HeatingQC.map(Exter_dict)
    df['CentralAir']=df.CentralAir.map(CentralAir_dict)
    df['Electrical']=df.Electrical.map(Electrical_dict)
    df['KitchenQual']=df.KitchenQual.map(Exter_dict)
    df['Functional']=df.Functional.map(Functional_dict)
    df['FireplaceQu']=df.FireplaceQu.map(Exter_dict)
    df['GarageQual']=df.GarageQual.map(Exter_dict)
    df['GarageCond']=df.GarageCond.map(Exter_dict)
    df['PavedDrive']=df.PavedDrive.map(CentralAir_dict)
    df['PoolQC']=df.PoolQC.map(Exter_dict)
    return df
```

```python
def changeattr(df):
    for col in df.columns:
        if(df[col].dtypes=='object'):
            df[col]=df[col].astype(str)
    return df
```

```python
def addextraattributes(df):
    now=dt.datetime.now()
    df['age_of_house']=now.year-df['YearBuilt']
    df['remodd_age']=now.year-df['YearRemodAdd']
    df['year_sold']=(df['YrSold'].astype(str)+(df['MoSold'].astype(str)).str.zfill(2)).astype(int)
    df['totalsqft']= df['TotalBsmtSF']+df['1stFlrSF']+df['2ndFlrSF']
    df['total_washroom']=df['BsmtFullBath']+df['BsmtHalfBath']+df['FullBath']+df['HalfBath']
    return df
  
```

## Scrub

```python
numPipeline=Pipeline([
 ('imputer',SimpleImputer(missing_values=np.nan,strategy='median')),
 #('poly',PolynomialFeatures(degree=2)),
 ('std_scaler', StandardScaler())
])

ordPipeline=Pipeline([
    ('imp',SimpleImputer(missing_values=np.nan,strategy='most_frequent'))
])

fullPipeline=ColumnTransformer([
    ('num',numPipeline,numcols),
    ("ord",ordPipeline,ordcols),
    ("cat",ce.BinaryEncoder(),catcols),
])
```

```python
X=traindf[analysiscols]
Y=traindf[target]
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=.50,random_state=0,shuffle=True)
```

```python
Xholdtrain,Xholdtest,Yholdtrain,Yholdtest=train_test_split(Xtest,Ytest,test_size=.20,random_state=0,shuffle=True)
```

## Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor
```

```python
Xtrainprep=fullPipeline.fit_transform(Xtrain)
```

```python
#Xtrainprep=KernelPCA(kernel='rbf').fit_transform(Xtrainprep)
```

```python
models=[LinearRegression(),
        Lasso(alpha=1,fit_intercept=False),
        ElasticNet(alpha=.5,l1_ratio=.5,fit_intercept=False),
        KNeighborsRegressor(),
        SVR(kernel='rbf'),
        SVR(kernel='sigmoid'),
        SVR(kernel='poly'),
        GradientBoostingRegressor(random_state=42),
        RandomForestRegressor(random_state=42),
        AdaBoostRegressor(random_state=42)]
```

```python
result_model={}
for model in models:
    result_model[model]=np.mean(np.sqrt(-cross_val_score(model,Xtrainprep,Ytrain,scoring='neg_mean_squared_error',cv=5)))
print(result_model)
```

## Tuning

```python
def modeltuning(reg,paramgrid,Xtrainprep,Ytrain):
    grid_search=GridSearchCV(reg(random_state=42),cv=5,param_grid=paramgrid,
                             scoring='neg_mean_squared_error',verbose=2,n_jobs=-1,return_train_score=True)
    grid_search.fit(Xtrainprep,Ytrain)
    model=grid_search.best_estimator_
    cvresult=pd.DataFrame(grid_search.cv_results_)
    cvresult['sqrt_mean_test_score']=np.sqrt(-cvresult['mean_test_score'])
    cvresult['sqrt_mean_train_score']=np.sqrt(-cvresult['mean_train_score'])
    result=cvresult[['params','sqrt_mean_test_score','sqrt_mean_train_score']]
    return result,model
```

```python
ada_params = [{'n_estimators': [1000],'learning_rate' : [.5,0.75] }]
ada_result,ada_model=modeltuning(AdaBoostRegressor,ada_params,Xtrainprep,Ytrain)
```

```python
elastic_params=[{'alpha':np.linspace(.3,1,10),'l1_ratio':np.linspace(0,1,10)}]
el_result,el_model=modeltuning(ElasticNet,elastic_params,Xtrainprep,Ytrain)
```

```python
rf_param=[{'n_estimators': [800,900,1000], 'max_features':['log2','sqrt']}]
rf_result,rf_model=modeltuning(RandomForestRegressor,rf_param,Xtrainprep,Ytrain)
```

```python
gb_param=[{'alpha':[.9],'learning_rate':[.09,.05,.025,.001],'n_estimators': [1000,1500,2000], 
           'min_impurity_decrease':[1e-5,1e-6],
           'max_features':[2,3,4,'log2'],'warm_start':[True]}]
gb_result,gb_model=modeltuning(GradientBoostingRegressor,gb_param,Xtrainprep,Ytrain)
```
